```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import tensorflow_datasets as tdfs
```

```python
#reading the input from pandas
GPA= pd.DataFrame(pd.read_excel(r"./mld.xlsx"))
```

```python
#forming the features
X=GPA[['Subject', 'Marks', 'Difficulty']]
Y=GPA[['Grade']]
plt.plot(X,Y)
```

```python
#trianing the model
model=LinearRegression()
model.fit(X,Y)
```

```python
#testing it
y_predict=model.predict([[0, 82, 2]])
np.floor(y_predict).astype(int)[0][0]
```
